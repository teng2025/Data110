```python
import matplotlib.pyplot as plt
import numpy as np

crime_categories = ["Assault", "Burglary", "Larceny", "Car Theft", "Murder", "Rape", "Robbery"]
rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
changes = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

# Create bar chart
plt.bar(crime_categories, rates_2023, label="Offenses (2023)", color='b')

# Add changes as text annotations on top of the bars
for i, change in enumerate(changes):
    plt.text(i, rates_2023[i] + 10, f"{change:+.1f}%", ha='center', va='bottom', color='black')  # Adjust positioning as needed

plt.xlabel("Offense")
plt.ylabel("Offenses per 100,000 People")
plt.title("Crime Data Visualization (2022-2023)")
plt.grid(axis='y')
plt.legend()
plt.show()

# Crime categories and their  data
crime_categories = ["Aggravated Assault", "Burglary", "Larceny", "Motor Vehicle Theft", "Murder", "Rape (revised definition)", "Robbery"]
rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
changes = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

# Sort the data in ascending order
sorted_indices = np.argsort(rates_2023)
crime_categories = [crime_categories[i] for i in sorted_indices]
rates_2023 = [rates_2023[i] for i in sorted_indices]
changes = [changes[i] for i in sorted_indices]

# Provide unique color for each crime
num_categories = len(crime_categories)
colormap = plt.cm.get_cmap('viridis', num_categories)

# horizontal bar chart with unique colors
fig, ax = plt.subplots()
for i, category in enumerate(crime_categories):
    ax.barh(category, rates_2023[i], color=colormap(i), label=category)

# text next to the bars
for i, change in enumerate(changes):
    ax.text(rates_2023[i] + 10, i, f"{change:+.1f}%", va='center', color='black')

# Customize the plot
ax.set_xlabel("Offenses per 100,000 People")
ax.set_ylabel("Offense")
ax.set_title("Crime Data Visualization (2023)")

plt.tight_layout()
plt.show()

#1 Why does this bar graph make more sense than Task 1's bar graph?
#It's easier for the audience to perceive the gradual changes in number of crimes. This is because of the descending size of the bar graphs.
#Additionally, the change over time is more insightful than just seeing the total numbers because it better explains the purpose of the data.
#If the percentage of change was disregarded then the audience would be less familiar with the transformation of crime rates.
```
