```python
# @title
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# View the first few rows of the dataset
titanic.head()

#TASK 1
#Histogram
plt.style.use('ggplot')
plt.figure(figsize=(10, 10))
plt.hist(titanic['fare'], bins=60, color='skyblue',edgecolor='white')
plt.title('Distribution of Titanic Passengers by Fare')
plt.xlabel('Fare')
plt.ylabel('Passengers')
plt.grid(axis='x')
plt.xticks([0, 100, 200, 300, 400, 500, 600 ])
plt.tick_params(axis='x', colors='gray')  # Set x-axis tick color
plt.show()

#TASK 1
#The Titanic dataset showcases the occupants of the Titanic. Additionally, their general info is provided:
#age, sex, fare, etc. The most important piece of information is whether someone lived or not. Lastly, there's
#a large amount of data, spanning from 1 - ~ 900.
#KDE
sns.kdeplot(titanic['fare'], color='red', linewidth=2, fill=True, alpha=0.2, bw_adjust=0.5)
plt.title('Density Plot of Titanic Passengers by Fare')
plt.xlabel('Fare')
plt.ylabel('Passengers')
plt.grid(axis='x')
plt.xticks([0, 100, 200, 300, 400, 500, 600 ])
plt.tick_params(axis='x', colors='gray')  # Set x-axis tick color
plt.show()

#TASK 2
#KDE
plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic, x='fare', hue='survived', fill=True, common_norm=False)

#Customize plot
plt.title('Fare Distribution of Survivors vs. Non-Survivors')
plt.xlabel('Fare')
plt.ylabel('Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

#TASK 3
# Create a figure with subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 12))

#TASK 4
# Create a figure with subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 12))

# Loop through passenger classes (1, 2, 3)
for i, pclass in enumerate([1, 2, 3]):
    # Filter data for current passenger class
    class_data = titanic[titanic['pclass'] == pclass]

    # Histogram for Fare Distribution
    sns.histplot(data=class_data, x='fare', hue='survived', bins=20,
    ax=axes[i, 0], multiple='stack', kde=False, legend=False)
    axes[i, 0].set_title(f'Fare Distribution for Pclass {pclass}')
    axes[i, 0].set_xlabel('Fare')
    axes[i, 0].set_ylabel('Count')
    axes[i, 0].grid(axis='x')

# KDE plot for Fare
sns.kdeplot(data=class_data, x='fare', hue='survived', fill=True, common_norm=False,
ax=axes[i, 1], legend=(i == 0))
axes[i, 1].set_title(f'Fare for Pclass {pclass}')
axes[i, 1].set_xlabel('Fare')
axes[i, 1].set_ylabel('Passengers')
axes[i, 1].grid(axis='x')

# Add a legend
handles, labels = axes[0, 1].get_legend_handles_labels()
fig.legend(handles, labels, title='Survived', loc='upper right')

plt.tight_layout()  # Adjust subplot spacing
plt.show()

#TASK 5
# Create the KDE plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.kdeplot(data=titanic, x='age', hue='survived', fill=True, common_norm=False)

# Customize the plot
plt.title('Age Distribution of Survivors vs. Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])  # Customize legend labels
plt.show()

#TASK 6
```
