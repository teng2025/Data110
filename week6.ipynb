```python
"""
# Palmer Penguins Data Analysis

"""

# Install packages
!pip install seaborn pandas matplotlib

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Inspect dataset
df.head()

# Remove rows with missing values
df = df.dropna()

# ---
# Boxplot: Body Mass by Species
# ---

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='body_mass_g', palette='Set2')
plt.title("Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()

# ---
# Violin Plot: Flipper Length by Species
# ---

plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='species', y='flipper_length_mm', palette='pastel')
plt.title("Flipper Length by Species")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.show()

# ---
# Ridgeline Plot (KDE): Bill Length by Island
# ---

plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='bill_length_mm', hue='island', fill=True, alpha=0.5)
plt.title("Bill Length Distribution by Island")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Density")
plt.show()

"""
## Interpretation Questions:

### Boxplot (Body Mass by Species)

1. Which species has the largest median body mass?
2. Which species has the widest range in body mass?
3. Are there any outliers? If so, which species have them?

The species with the largest median body mass is Gentoo, which also has the widest range in body mass.
Yes the outliers are mainly in the Chinstrap species.

### Violin Plot (Flipper Length by Species)

1. Which species has the most symmetrical distribution?
2. Which species has the widest spread of flipper lengths?
3. How do the quartiles (white lines in the plot) compare for each species?

The most symmetrical distribution is in Chinstrap species. Gentoo has the widest spread of flipper lengths.
The quartiles show that Gentoo penguins have longer flippers than the other species.

### Ridgeline Plot (Bill Length by Island)

1. Which island has the smallest range of bill lengths?
2. Which island has the most uniform distribution?
3. What differences do you notice between the islands in terms of bill length?

Biscoe Island has the smallest range of bill lengths. Dream Island has the most uniform distribution.
Torgersen Island has a sharper peak, showing a more concentrated bill length range compared to others.
```
