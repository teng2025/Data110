```python
import matplotlib.pyplot as plt
import numpy as np

#1
plt.style.use('seaborn-v0_8-whitegrid')
print(plt.style.available)

#2
# Define the function
def f(x):
    return x**2 - 2*x + 1

# Create the x-values
x = np.linspace(-2, 4, 100)  # 100 points between -2 and 4

#3
# Calculate the y-values
y = f(x)

# Create the plot
plt.plot(x, y)
# Display the plot
plt.show()
# Add labels and title
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = x - 2x + 1")

# Define the functions
def f(x):
    return x**3
def g(x):
    return np.exp(x)
def h(x):
    return np.log(x)

#3
# Create the x-values
x = np.linspace(0.1, 5, 100)  # 100 points between 0.1 and 5 (log(x) needs x > 0)

# Calculate the y-values for each function
y1 = f(x)
y2 = g(x)
y3 = h(x)

# Create the plot
plt.plot(x, y1, label='f(x) = x', color='blue')
plt.plot(x, y2, label='g(x) = e', color='red')
plt.plot(x, y3, label='h(x) = log(x)', color='green')

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of Three Functions")

# Add a legend
plt.legend()

# Define the functions
def f(x):
    return x**2
def g(x):
    return x**3
def h(x):
    return x**4
def i(x):
    return x**5

# Create the x-values
x = np.linspace(-2, 2, 100)

# Calculate the y-values for each function
y1 = f(x)
y2 = g(x)
y3 = h(x)
y4 = i(x)

#4
# Create the plot with different colors and line styles
plt.plot(x, y1, label='f(x) = x', color='blue', linestyle='-')       # Color name
plt.plot(x, y2, label='g(x) = x', color='g', linestyle='--')        # Short color code
plt.plot(x, y3, label='h(x) = x', color='#FFA500', linestyle=':')  # Hex color code
plt.plot(x, y4, label='i(x) = x', color=(0.8, 0.2, 0.5), linestyle='-.') # RGB tuple

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of Four Functions")

# Add a legend
plt.legend()

#5
#Definefunction
def f(x):
    return 1/(x**2 + 1)

# x-values for  function
x_new = np.linspace(-10, 10, 400)

# y-values for function
y_new = f(x_new)

plt.plot(x_new, y_new, label='f(x) = 1/(x + 1)', color='red', linestyle='--')

plt.show()

#6
# x-values for x  0
x = np.linspace(0, 5, 100)  # Adjust range as needed

# Define the functions
def f(x):
    return x
def g(x):
    return x**(1/3)

# Calculate y-values
y1 = f(x)
y2 = g(x)

# Create the plot
plt.plot(x, y1, label='f(x) = x')
plt.plot(x, y2, label='g(x) = x^(1/3)')

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of f(x) = x and g(x) = x^(1/3)")

# Add a legend
plt.legend()

# Display the plot
plt.show()

#7
# Create 50 random points for x and y
x = np.random.rand(50)
y = np.random.rand(50)

# Create the scatter plot using plt.plot()
plt.plot(x, y, 'o')  # 'o' marker for circles

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot using plt.plot()")

# Display plot
plt.show()

#8
# Create data for the function f(x) = x*sin(x)
x = np.linspace(0, 4*np.pi, 100)  # 100 points between 0 and 4
y = x * np.sin(x)

# Create the scatter plot with a line connecting the points
plt.plot(x, y, marker='o', linestyle='-', markersize=8, linewidth=2,
         markerfacecolor='blue', markeredgecolor='black', markeredgewidth=1.5)

# Add labels and title
plt.xlabel("x")
plt.ylabel("f(x) = x*sin(x)")
plt.title("Scatter Plot of x*sin(x)")

# Display plot
plt.show()

#9
# Define function
def f(x):
    return x * np.cos(x)

# Create x-values for the line plot
x_line = np.linspace(-2*np.pi, 2*np.pi, 100)

# Calculate y-values for the line plot
y_line = f(x_line)

# Create x-values for scatter points at /4 intervals
x_scatter = np.arange(-2*np.pi, 2*np.pi, np.pi/4)

# Calculate y-values for scatter points
y_scatter = f(x_scatter)

# Create the line plot
plt.plot(x_line, y_line, color='blue', linestyle='-', label='f(x) = x*cos(x)')

# Add scatter points
plt.scatter(x_scatter, y_scatter, color='red', marker='o', label='Scatter Points')

# Add labels and title
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = x*cos(x) with Scatter Points")

# Add legend
plt.legend()

# Display the plot
plt.show()

#10
# 'bmh' style from Task One
plt.style.use('bmh')

# Define the functions
def i(x):
    return np.arctan(x)  # Inverse tangent function
def j(x):
    return 1/(1 + np.exp(-x))  # Logistic function

# Create x-values
x = np.linspace(-5, 5, 100)

# Calculate y-values
y1 = i(x)
y2 = j(x)

# Create the plot
plt.plot(x, y1, label='i(x) = arctan(x)', color='navy', linestyle='--', linewidth=2)
plt.plot(x, y2, label='j(x) = 1/(1 + e^(-x))', color='darkorange', linestyle='-', linewidth=2)

# scatter points for arctan(x) at specific intervals
x_scatter = np.arange(-5, 6, 2)
y_scatter = i(x_scatter)
plt.scatter(x_scatter, y_scatter, color='navy', marker='o', s=50, edgecolors='black') # Scatter points for arctan

# custom limits
plt.xlim(-5, 5)
plt.ylim(-2, 2)

#labels a nd title
plt.xlabel("x")
plt.ylabel("Function Values")
plt.title("Custom Plot: Arctan and Logistic Functions")

# legend
plt.legend()

# Display the plot
plt.show()

#11
# Create data for the function f(x) = x*sin(x)
x = np.linspace(0, 4*np.pi, 100)  # 100 points between 0 and 4
y = x * np.sin(x)

# Create the scatter plot with a line connecting the points
plt.plot(x, y, marker='o', linestyle='-', markersize=8, linewidth=2,
         markerfacecolor='blue', markeredgecolor='black', markeredgewidth=1.5)

# Add labels and title
plt.xlabel("x")
plt.ylabel("f(x) = x*sin(x)")
plt.title("Scatter Plot of x*sin(x)")

# Display plot
plt.show()
```
